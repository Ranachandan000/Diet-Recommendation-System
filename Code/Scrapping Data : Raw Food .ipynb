# 01

```python
base_url = "http://218.248.6.43:8080/CountWhatYouEat/vitamincontentfoods.do?"

all_data = []

for page_number in range(1, 8):  
    url = base_url + "d-16460-p=" + str(page_number)

    response = requests.get(url)

    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", {"id": "ite"})

    data = []
    headers = []
    rows = table.find_all("tr")
    for i, row in enumerate(rows):
        if i == 0:
            headers = [header.text for header in row.find_all("th")]
        else:
            data.append([cell.text for cell in row.find_all("td")])

    all_data.extend(data)

df = pd.DataFrame(all_data, columns=headers)

print(df)
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://218.248.6.43:8080/CountWhatYouEat/carbohydrates.do?d-16460-p=1"

response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(html_content, 'html.parser')

data = []

table_rows = soup.find_all('tr')
for row in table_rows:
    foodname_input = row.find('input', {'type': 'hidden', 'name': 'fn', 'id': 'foodname'})
    if foodname_input:
        foodname = foodname_input.find_next('input')['value']
        value_td = row.find('td', text=True)
        value = value_td.text
        data.append((foodname, value))

print(data)
```
